### rename the class number

```python
def rename_class(source_file: str, dest_file: str):

    with open(source_file, "r", encoding="utf-8") as file:
        row = file.readlines()
        # row = [r.strip("\n") for r in row]
        new_row = []
        for r in row:
            new_row.append(r.replace("254 ", "0 "))
            
        with open(dest_file, "w+", encoding="utf-8") as write_file:
            write_file.writelines(new_row)
```

```python
train_mask_path = os.listdir(TRAIN_MASK_PATH)
train_sat_path = os.listdir(TRAIN_SAT_PATH)
test_mask_path = os.listdir(TEST_MASK_PATH)

train_target_mask_path = "data/splitted/deepglobe/labels/train_renamed"
test_target_mask_path = "data/splitted/deepglobe/labels/val_renamed"
```
